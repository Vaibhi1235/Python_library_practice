```python
def Patient_pulse(one):
    if one>100 and one <125:
        New_pulse=one-10
        return New_pulse
    else:
        return  one
```

```python
update_file['Pulse']=update_file['Pulse'].apply(Patient_pulse)
update_file.tail()
```

```python
def calories_burn_limit(value):
    if value>220 and value <290:
        new=value
        return value
    elif value>290 and value <350:
        new=value-80
        return new
    elif value >350 and value<400:
        new =value-120
        return new
    else:
        return value-160
```

```python
update_file['patient_update_calories']=update_file['Calories'].apply(calories_burn_limit)
update_file.head(10)
```
